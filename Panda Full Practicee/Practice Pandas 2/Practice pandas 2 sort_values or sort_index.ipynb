## Sorting datafram
- axis = 0 --> by columns
- axis = 1 --> by rows

```python
import pandas as pd
import numpy as np
import os
os.listdir()
```

```python
pd.__version__
```

```python
emp_rec = pd.read_excel("Employee_Records.xlsx")
emp_rec
```

```python
emp_rec.sort_values(["Age in Yrs"] ,axis=0)
```

```python
emp_rec.sort_index(axis=1)
```

```python
emp_rec.info()
```

```python
nw = emp_rec.loc[0:,["First Name", "Age in Yrs"]].head(10)
nw
```

```python
nw.sort_values("Age in Yrs", ignore_index=True)
```

```python
nw.sort_values(["Age in Yrs", "First Name"], ignore_index=True)
```

```python
sort_name_wise = nw.sort_values(["First Name"])
```

```python
sort_name_wise
```

```python
sort_name_wise.reset_index()
```

```python
sort_name_wise.reset_index(drop=True)
```
