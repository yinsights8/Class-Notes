### 2. .concat() :-
- The concat() function is used to concatenate pandas objects along a particular axis with optional set logic along the other axes.
- It concatinats the Dataframe on the Lables of index and columns

##### Syntax 
    - pd.concat(objects=[df1,df2..dfn], axis=0/1, join = "outer"/ "inner")
- 1. join = "inner"
    - This parameter will concatenat the the ColumnLables which are intersecting in both the dataframes(df1,df2) and remaining columnLables will be droped. (axis=0)
    
    - (axis=1) at axis=1  This parameter will concat the indexes which ara intersecting or common  in both the IndexLables.
- ###### Note :- if No indexLables are common in both dataframes then no Concatination will be there..
 
 
 -
- 2. join = "outer"
    - bydefault it is set to "outer"
    - it will concatenats all the ColumnsLables and IndexLables from both the dataframes at axis=1 and axis=0


###### NOTE:-  (Intersect - Common  ColumnLables  and  IndexLables  in both df1 and df2)

### 3. join()
- join() method is use to join the two dataframes.

#### - Syntax
- Syntax:-
    - df_name1.join(df_name2, on="IndexLable", how = "inner"/ "outer"/ "left"/ "right")
    -
    - "left":- joins all the row elements from df1(left) as well as common / intersecting elements from df1 (left) and df2 (right).
    -
    - "rigth" :- joins all the row elements from df1(right) as well as common / intersecting elements from df1 (left) and df2 (right).


- Return Value
    - A new DataFrame, with the updated result.

    - This method does not change the original DataFrame.
    
###### NOTE:- Condition is, indexes must be same in both the DataFrames in order to perform "left", "right", "inner", "outer"

### 3. merge(left_df, right_df, how = "inner" / "outer")
- pd.merge() method is use to merge the left and right dataframe.
#### Syntax
- Syntax:-
    - pd.merge(left_df, right_df, on="ColumnLables", how="inner" / "outer")

### Differnce betweenb join() and merge()
- Both join and merge can be used to combines two dataframes but the join method combines two dataframes on the basis of their indexes whereas the merge method is more versatile and allows us to specify columns beside the index to join on for both dataframes.
