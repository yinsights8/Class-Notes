## 1. Supervised Learning (SL):
- Supervised learning, also known as supervised machine learning, is a subcategory of machine learning and artificial intelligence. 
-It is defined by its use of labeled datasets to train algorithms that to classify data or predict outcomes accurately.
- it is a method in which we teach machine using labled data



- Example :- we will pass some historical data to machine and we will teach the machine. based on the previous data our machine will predict.

- 1. In Supervised Machine Learning, Our data will be labeled
- 2. Independent and Dependent Variables will be present in Data

### Methods in SL
- 1. Regression:
    
    - 1. Continuous Target column (Nature of Our data will be Continuous)
    ex. Price, Score
    
- 2. Classification:
    
    - 1. Binary Classification (True or False, Yes or No, Spam or Not Spam, 0 or 1, Approved or Declined)
    - 2. Multiclass Classification (High or Medium or Low)
      

### applications of SL

    - Object recognition for the vision
