### Logloss Function curves

```python
import numpy as np

import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
p = np.arange(0,1,0.01)
```

```python
## y = 1
plt.figure(figsize=(8,7),facecolor="white")

plt.plot(p, -1 * np.log(p))
plt.title("y=1",fontsize=20)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
## when y = 0
plt.figure(figsize=(8,7),facecolor="white")

plt.plot(p, -1 * np.log(1-p))
plt.title("y=0",fontsize=20)
plt.xlabel("X-axis")
plt.ylabel("log(1-Yi)")
```

```python
plt.figure(figsize=(8,7),facecolor="white")

# y = 1
plt.plot(p, -1 * np.log(p))

# y = 0
plt.plot(p, -1 * np.log(1-p), "r")



plt.xlabel("X-axis")
plt.ylabel("Yi")
```
