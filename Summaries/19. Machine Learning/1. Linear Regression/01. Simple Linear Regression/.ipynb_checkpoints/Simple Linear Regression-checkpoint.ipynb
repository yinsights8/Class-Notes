```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url =r"https://raw.githubusercontent.com/chasinginfinity/ml-from-scratch/master/01%20Linear%20Regression%20using%20Least%20Squares/data.csv"

df = pd.read_csv(url)
# df.head()
```

```python
### data spliting

x = df.iloc[:, :-1].values
y = df.iloc[:,-1].values
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
## calculating a best M and C value

num = 0
den = 0
for i in range(len(df)):
    num += (x[i] - x_mean)*(y[i] - y_mean)
    den += (x[i] - x_mean)**2
    
m = num/den
c = y_mean - m*x_mean

print(f" Best M and C values are \nM = {m}, \nC = {c}")

## y = 1.28735737x + 9.90860619
```

```python
## predicting values
# Y = Mx + c

y_pred = m*x + c
```

```python
plt.scatter(x,y)
plt.scatter(x,y_pred, color="red")
# plt.plot(x, y_pred, color="red")
plt.show()
```

```python
plt.scatter(x,y)
# plt.scatter(x,y_pred, color="red")
plt.plot([max(x), min(x)], [(max(y_pred)),(min(y_pred))], color="red")
plt.show()
```
