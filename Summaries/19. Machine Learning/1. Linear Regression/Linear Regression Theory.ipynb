## How we can check for normality of Residuals

```python
- residuals should be normally distributed 
- residuals = Ya - Yp

- we can check for the Normality by following tests

    - 1. Visiulization tools (import libraris form seaborn)
        - 1. kdeplot 
        - 2. histplot
    
    - 2. QQ-plot (Best method to check Normality) >>  Quantile-Quantile plot
            - import QQ-plot tests
                - import statsmodels.api as sm
            
    - 3. Skewness of residulas :-
            - import skewness of residuals
                - import scipy.stats import skew
            1. Positive Skewness >> Tail of the Distribution is longer towards right hand side
            2. Negative Skewness >> Tail of the Distribution is longer towards left hand side

            Symmetrical Data >> Bell Shape Curve

            skew > 1.0 : Highly positive Skewed Data 
            skew < -1.0 : Highly Negative Skewed Data
    
    - 4. Hypothesis Test (use any one method from bellow) 
        - import Hypothesis tests
            - from scipy.stats import shapiro, kstest, normal test
            
            - 1. Shapiro Test
            - 2. KSTest  (Kolmogorov - Smirnov - Test)
            - 3. Normal Test
            
            shapiro/kstest/normaltest tests p_value >= 0.05 >> we are in Null Hypothesis 
                                                        All Values are Normally Distributed.
                                                        (Null Hypothesis is True)
                
            shapiro/kstest/normaltest tests p_value < 0.05 >> we are in Alternate Hypothesis 
                                                             All Values are not Normally Distributed.
                                                             (Alternate Hypothesis is True)
```

### Types of Hypothesis Test

```python
Types of Hypothesis :
    
    1. Null Hypothesis >> We are accepting Null Hypothesis
                          Null Hypothesis True
        
    2. Alternate Hypothesis >> Alternate Hypothesis is True
                               We are Acceptiong Alternate Hypothesis
                               We are Rejecting Null Hypothesis 
                               Null hypothesis will be False
                
1. Null Hypothesis >> Data is Normally Distributed
2. Alternate Hypothesis >> Data is not Normally Distributed

p_val >> 0 to 1
Significant Value of P :   p_val == 0.05
    
    
p_val >= 0.05 : Null hypothesis is True
p_val < 0.05  : Null Hypothesis is False
                Alternate Hypothesis is True
        
        
p_val >= 0.05 : Data is Normally Distributed
p_val < 0.05  : Data is not Normally Distributed
    
NOTE :- remember the universe and coin toss Example
```

# Feature Scaling 
- This is the most important transformation method to apply on our dataset.
- Machine Learning algorithms dont perform well when the input numerical attributes have very different scales
- Feature Scaling is a technique to standardize the independent features present in the data in a fixed range.

-  A significant issue is that the range of the variables may differ a lot. Using the original scale may put more weights on the variables with a large range. In order to deal with this problem, we need to apply the technique of features rescaling to independent variables or features of data in the step of data pre-processing.

- Unlike min-max scaling, standardization does not bound values to a specific range,
        - e.g., neural networks often expect an input value ranging from 0 to 1
- However, standardization is much less affected by outliers.

###### NOTE :- The goal of applying Feature Scaling is to make sure features are on almost the same scale so that each feature is equally important and make it easier to process by most ML algorithms
######  NOTE :- If we do not do the feature scaling Then the performance of Our model may not be well, Because the most dominant feature of the given dataset our Model will ignore the previous features which are important but less dominant

## Bias - Variance and Trade-Of

- if the machine learning model is not accurate, it can make predictions errors, and these prediction errors are usually known as Bias and Variance. 
- In machine learning, these errors will always be present as there is always a slight difference between the model predictions and actual predictions

- Bias :- 
    - it can be defined as an inability of machine learning algorithms such as Linear Regression to capture the true relationship between the data points.

- Variance
    - The variance would specify the amount of variation in the prediction if the different training data was used. In simple words, **variance tells that how much a random variable is different from its expected value**. 
     
    - Ideally, a model should not vary too much from one training dataset to another, which means the algorithm should be good in understanding the hidden mapping between inputs and output variables. Variance errors are either of low variance or high variance.

    - Low variance means there is a small variation in the prediction of the target function with changes in the training data set. At the same time, High variance shows a large variation in the prediction of the target function with changes in the training dataset.

     - A model that shows high variance learns a lot and perform well with the training dataset, and does not generalize well with the unseen dataset. As a result, such a model gives good results with the training dataset but shows high error rates on the test dataset.

### - Bias
    - it is An Error of the Training data points
    
### - Variance 
    - It is An Error of the test data points
- variance is a measure of how far observed values differ from the average of predicted values
- The variance of Y is equal to the variance of predicted values plus the variance of the residuals.
- The goal is to have a value that is low.



## - Covariance
    - Covariance shows you how the two variables differ. 
- Covariance indicates the relationship of two variables whenever one variable changes. If an increase in one variable results in an increase in the other variable, both variables are said to have a positive covariance. Decreases in one variable also cause a decrease in the other.
    - (X , & Y  = +ve covariance)
    
    - (X , & Y  = -ve covariance)
    
###### NOTE :- It tells you that what is the direction of the variables is, But is dosent show how much it is +vely correlated  and  how much it is -ve correlated 

## - Correlation (Pearson cofficient of correlations or cofficient of corelation)
    - correlation shows you how the two variables are related
- it shows how strongly variables is corelated (X and Y)
