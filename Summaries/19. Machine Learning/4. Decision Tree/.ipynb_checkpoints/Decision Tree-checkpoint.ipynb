```python
import pandas as pd
import numpy as np
```

```python
data = {"X":np.arange(1,15), 
        "Y":np.array([1,1.2,1.4,1.1,1,5.5,6.1,6.7,6.4,6,6,3,3.2,3.1])}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df = df.sort_values("X")
```

```python
df
```

```python
def Average(df, x):
    avgs = []
    values_A = []
    values_B = []
    for i in range(len(x)-1):
        avg = (x[i] + x[i+1]) / 2
        avgs.append(avg)
        
    for i in range(len(avgs)-1):
        part_A = df[df["X"] <= avgs[i]].values
        part_B = df[df["X"] > avgs[i+1]].values
        values_A = np.append(arr=values_A,values=part_A)
        values_B = np.append(arr=values_B, values=part_B)
    return values_A
#     return avgs

Average(df, df["X"])
```
