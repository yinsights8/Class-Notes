```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from io import StringIO

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import warnings 
warnings.filterwarnings("ignore")
```

```python
d1 = {"Score" : np.arange(1,40),
       'result': np.zeros(shape=(39), dtype=int)}

d2 = {"Score" : np.arange(40,101),
       'result': np.ones(shape=(61), dtype=int)}
df2 = pd.DataFrame(d1)

df = df2.append(pd.DataFrame(d2), ignore_index=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
X_train = df.loc[:, df.columns != 'result']
Y_train = df.result
```

```python
## Calculating Entropy for Result
# E(s) = -P(yes) * log2(P(yes)) - P(no) * log2(P(no))
# 0 = fail 1 = pass
```

```python
def entropy(target_col):
    element, count = np.unique(target_col, return_counts=True)
    
    for e,c in zip(element, count):
        print("Node has {} elements of Class {}".format(c, e))
    entropy = np.sum([-(count[i] / np.sum(count)) * np.log2(count[i]/ np.sum(count)) 
                      for i in range(len(element))])
    return entropy
```

```python
entropy_before_split = entropy(df.result)
print ("Initial Entropy = {}".format(entropy_before_split))
```

```python
split_values = np.arange(1.5,100,1)
split_values
```

```python
left_node_data = df.loc[df.Score < split_values[0]]
right_node_data = df.loc[df.Score > split_values[0]]
```

```python
left_node_data
```

```python
right_node_data
```

```python
def calc_info_gain_numeric(df, left, right, entropy_before_split, split_values):
    print("Split Value = {} \n".format(split_values))
    total_elements = df.shape[0]
    print("Left Node")
    ent_left = entropy(left)
    print("Entropy of Left Node {} \n".format(ent_left))
    print("Right Node")
    ent_right = entropy(right)
    print("Entropy of Right Node {} \n".format(ent_right))
    weighted_entropy = ((left.shape[0] / total_elements) * ent_left) + (
        (right.shape[0] / total_elements) * ent_right
    )
    info_gain = entropy_before_split - weighted_entropy
    print("Info gain at split {} is {} \n".format(split_values, info_gain))
    print("*"* 50)
    return info_gain
```

```python
calc_info_gain_numeric(
    df, left_node_data.result, right_node_data.result, entropy_before_split, split_values[0]
)
```

```python
info_gains = []
for sp_value in split_values:
    left_data = df.loc[df.Score < sp_value]
    right_data = df.loc[df.Score > sp_value]
    info_gains.append(
        calc_info_gain_numeric(
            df, left_data.result, right_data.result, entropy_before_split, sp_value
        )
    )
```

```python
(max(info_gains))
```

```python
clf = DecisionTreeClassifier(criterion='entropy')
```

```python
clf.fit(X_train,Y_train)
```

```python
plot_tree(decision_tree=clf, feature_names=X_train.columns,class_names=["Fail", "Pass"])
```
