##### Check for Normality Assumption

```python
## arr1
plt.figure(figsize=(5,4))
sns.kdeplot(arr1, fill=True)

## As we the data is normally distributed for arr1
```

```python
## arr2
plt.figure(figsize=(5,4))
sns.kdeplot(arr2, fill=True)

## As we the data is not normally distributed for arr2, the curve is right skewed
```

```python
## Array1

from scipy.stats import shapiro, kstest, normaltest

stat, p_value = shapiro(arr1)
if p_value >= 0.05:
    print(f"Shapiro Test :-\nNull Hypothesis is True\nAll the data points are Normally distributed \n{p_value}")
else:
    print(f"Shapiro Test:-\nAlternate Hypothesis is True\nAll the Data Points are not normally distributed\n{p_value}")
    
print("*"*50,"\n")

stat, p_value = normaltest(arr1)
if p_value >= 0.05:
    print(f"Normal Test:- \nNull Hypothesis is True\nAll the data points are Normally distributed \n{p_value}")
else:
    print(f"Normal Test:-\nAlternate Hypothesis is True\nAll the Data Points are not normally distributed\n{p_value}")
    
print("*"*50,"\n")

stat, p_value = kstest(arr1,cdf="norm")
if p_value >= 0.05:
    print(f"KsTest:-\nNull Hypothesis is True\nAll the data points are Normally distributed \n{p_value}")
else:
    print(f"KsTest:-\nAlternate Hypothesis is True\nAll the Data Points are not normally distributed\n{p_value}")
```

```python
## Arr2
from scipy.stats import shapiro, kstest, normaltest

stat, p_value = shapiro(arr2)
if p_value >= 0.05:
    print(f"Shapiro Test :-\nNull Hypothesis is True\nAll the data points are Normally distributed \n{p_value}")
else:
    print(f"Shapiro Test:-\nAlternate Hypothesis is True\nAll the Data Points are not normally distributed\n{p_value}")
    
print("*"*50,"\n")

stat, p_value = normaltest(arr2)
if p_value >= 0.05:
    print(f"Normal Test:- \nNull Hypothesis is True\nAll the data points are Normally distributed \n{p_value}")
else:
    print(f"Normal Test:-\nAlternate Hypothesis is True\nAll the Data Points are not normally distributed\n{p_value}")
    
print("*"*50,"\n")

stat, p_value = kstest(arr2,cdf="norm")
if p_value >= 0.05:
    print(f"KsTest:-\nNull Hypothesis is True\nAll the data points are Normally distributed \n{p_value}")
else:
    print(f"KsTest:-\nAlternate Hypothesis is True\nAll the Data Points are not normally distributed\n{p_value}")
```

#### calculating z-score for arr1 

```python
mean_arr1 = np.mean(arr1)
std_arr1 = np.std(arr1)
print(f"Mean of arr1 :- {mean_arr1}")
print(f"std of arr1 :- {std_arr1}")

thrsh = 3
outliers = []
for val in arr1:
    z_score = (val - mean_arr1)/ std_arr1
    if np.abs(z_score) > thrsh:
        outliers.append(val)
        
print(f"Outliers :- {outliers}")
```

### calculating z-score for arr2

```python
mean_arr2 = np.mean(arr2)
std_arr2 = np.std(arr2)
print(f"Mean of arr2 :- {mean_arr2}")
print(f"std of arr2 :- {std_arr2}")

thrsh = 2 # there were no Outliers in 3 std
outliers = []
for val in arr2:
    z_score = (val - mean_arr2)/ std_arr2
    if np.abs(z_score) > thrsh:
        outliers.append(val)
        
print(f"Outliers :- {outliers}")
```

## IQR Method

```python
## Array 2
print("Array 2 :- ",array2,"\n")


Q1 = np.quantile(array2, 0.25)  # 25th Percentile
Q3 = np.quantile(array2, 0.75)  # 75th Percentile
print(f"Q1 = {Q1}\nQ3 = {Q3}\n")

IQR = Q3 - Q1
print(f"Inter Quantile Range = {IQR}\n")

lower_tail = Q1 - 1.5 * IQR
upper_tail = Q3 + 1.5 * IQR    ## Normal outliers

print(f"lower_tail = {lower_tail}\nupper_tail = {upper_tail}\n")
print(f"""If any point in an array is lower than "{lower_tail}" 
and greater than "{upper_tail}" considerd to be 'Extreme outliers'""")
```
