```python
df=pd.read_sql("""select emp_id,emp_name,e.dept_id,dept_name,m.mgr_name,P.PROJ_NAME,P1.TEAM_MEMBER_ID
from employee E
left join department D
ON E.dept_id=d.dept_id
LEFT JOIN 
manager M
ON E.MGR_ID=M.MGR_ID
LEFT JOIN 
project p
ON E.EMP_ID=P.TEAM_MEMBER_ID
LEFT JOIN 
PROJECT P1
ON M.MGR_ID=P1.TEAM_MEMBER_ID;
""",engine)
df
```

### to sql

```python
df.to_sql(name='join_df',con=engine,index=False)
```
