```python
## word2vec return the words which are having same contexts or properties or characteristics
# if the the vectors are close to -1 then they do not have  same context or properites
# if the the vectors are close to +1 then they do  have  same context or properites
```

```python
path = r"E:\DS CLASS\Notes\7.Month 7\Day - 128 NLP Word2vec\GoogleNews-vectors-negative300.bin.gz"
```

```python
from gensim.models import KeyedVectors
model = KeyedVectors.load_word2vec_format(path,binary=True)
```

```python
model["bad"]
```

```python
## 300 >> dimensions >> features >> properties
len(model['bad'])
```

```python
# both the words have same context, therefore similarity will be closer to +1
model.similarity('bad', 'good')
```

```python
model.similarity("bad","ugly")
```

```python
# the following command is showing the cossine similarity of the given word
model.most_similar("bad")
```

```python
model.most_similar('king')
```

```python
# following command is for odd man out
model.doesnt_match(["python","google","sql","deer"])
```

```python
model.doesnt_match(["python","sql","snake"])
```

```python
model.similar_by_word("python")
```
