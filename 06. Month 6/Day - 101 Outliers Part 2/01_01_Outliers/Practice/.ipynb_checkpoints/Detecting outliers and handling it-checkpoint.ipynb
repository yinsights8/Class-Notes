```python
import numpy as  np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Detecting Outliers by following methods

###  Detecting Outliers by Z_score

```python
array1 = np.random.randint(20,40, size=(20))
```

```python
array1
```

```python
array2 = np.array([-25, 25, 39, -39, 29, 25, 315, 21, 31, 22, 39, 21, 34, 36, 35, 122, 31,
       23, 225, 39])
```

```python
## array 1
sns.set_theme(color_codes=True)
plt.figure(figsize=(5,4))
sns.kdeplot(array1, fill=True)
```

```python
## array 2
sns.set_theme(color_codes=True)
plt.figure(figsize=(5,4))
sns.kdeplot(array2, fill=True)
```

```python
### detecting Outliers by boxplot
sns.set_theme(color_codes=True)
plt.figure(figsize=(5,4))
sns.boxplot(array1)
```

```python
sns.set_theme(color_codes=True)
plt.figure(figsize=(5,4))
sns.boxplot(array2)
```

```python
### Detecting outliers by z Score

## Z_Score = X - Mean(X) / std

mean_x = np.mean(array1)
std_x = np.std(array1)

outliers = []
thre = 3
for x_val in array1:
    z_score = (x_val - mean_x) / std_x
    if np.abs(z_score) > thre:
        outliers.append(x_val)
        
print(outliers)
```

```python
## Array 2
### Detecting outliers by z Score
## Z_Score = X - Mean(X) / std

mean_x2 = np.mean(array2)
std_x2 = np.std(array2)

outliers = []
thre = 3 
for x_val in array2:
    z_score = (x_val - mean_x2) / std_x2
    if np.abs(z_score) > thre:
        outliers.append(x_val)
        
print("Outliers in array 2:- ",outliers)
```

### IQR Method
- IQR  >> Inter Quartile Range

```python
print(array1)
plt.figure(figsize=(4,3))
sns.boxplot(array1, palette="Set2")
```

```python
print(array2)
plt.figure(figsize=(4,3))
sns.boxplot(array2, palette="Set2")
```

```python
## Array 1 
print("Array 1 :- ",array1,"\n")


Q1 = np.quantile(array1, 0.25)  # 25th Percentile
Q3 = np.quantile(array1, 0.75)  # 75th Percentile
print(f"Q1 = {Q1}\nQ3 = {Q3}\n")

IQR = Q3 - Q1
print(f"Inter Quantile Range = {IQR}\n")

lower_tail = Q1 - 1.5 * IQR
upper_tail = Q3 + 1.5 * IQR    ## Normal outliers

print(f"lower_tail = {lower_tail}\nupper_tail = {upper_tail}\n")
print(f"""If any point in an array is lower than "{lower_tail}" 
and greater than "{upper_tail}" considerd to be 'Normal outliers' """)
```

```python
## Array 1 
print("Array 1 :- ",array1,"\n")


Q1 = np.quantile(array1, 0.25)  # 25th Percentile
Q3 = np.quantile(array1, 0.75)  # 75th Percentile
print(f"Q1 = {Q1}\nQ3 = {Q3}\n")

IQR = Q3 - Q1
print(f"Inter Quantile Range = {IQR}\n")

lower_tail = Q1 - 3 * IQR
upper_tail = Q3 + 3 * IQR    ## Extreme outliers

print(f"lower_tail = {lower_tail}\nupper_tail = {upper_tail}\n")
print(f"""If any point in an array is lower than "{lower_tail}" 
and greater than "{upper_tail}" considerd to be 'Extreme outliers'""")
```
