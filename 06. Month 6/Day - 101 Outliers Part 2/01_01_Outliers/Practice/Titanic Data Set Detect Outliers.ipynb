```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("titanic.csv")
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data["Age"].fillna(data["Age"].median(), inplace=True)
data["Gender"].fillna(data["Gender"].mode()[0], inplace=True)
```

```python
data.isna().sum()
```

```python
data.describe()
```

```python
## Outliers in Age Column
q1 = data["Age"].quantile(0.25)
q2 = data["Age"].quantile(0.50)
q3 = data["Age"].quantile(0.75)
print(f"Q1 = {q1}\nQ2 = {q2}\nQ3 = {q3}")
```

```python
# IQR = Q3 - Q1
IQR = q3 - q1
print(f"Inter Quantile Range :- {IQR}")

lower_tail = q1 - 1.5 * IQR
Upper_tail = q3 + 1.5 * IQR ## Normal outliers
print(f"\nRange for Normal Outliers :-\nLower tail :- {lower_tail}\nUpper tail :- {Upper_tail}")
```

```python
# IQR = Q3 - Q1
IQR = q3 - q1
print(f"Inter Quantile Range :- {IQR}")

lower_tail = q1 - 3 * IQR
Upper_tail = q3 + 3 * IQR ## Extreme outliers
print(f"\nRange for Normal Outliers :-\nLower tail :- {lower_tail}\nUpper tail :- {Upper_tail}")
```
