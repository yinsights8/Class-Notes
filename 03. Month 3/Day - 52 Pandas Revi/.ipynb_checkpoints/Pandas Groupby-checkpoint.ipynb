# Pandas Groupby
- df.groupby()
- df.groupby.first()
- df.groupby.get_group()

```python
import pandas as pd
import numpy as np
```

```python
movies_df = pd.read_csv("movies_data.csv")
movies_df.head()
```

```python
movies_df.info()
```

```python
movies_df.groupby("genre")
```

```python
movies_df.groupby("genre").first()
```

```python
movies_df.groupby("genre").get_group("Thriller")
```

```python
movies_df.groupby("genre").get_group("Drama")
```

```python
drama_df = movies_df.groupby("genre").get_group("Drama")
drama_df.reset_index()
```

```python
drama_df = movies_df.groupby("genre").get_group("Drama")
drama_df.reset_index(drop=True)
```

```python
drama_df.to_csv("DramaMovie.csv")
```

```python
movies_df.head()
```

```python
movies_df.groupby(["genre","content_rating"]).get_group(("Crime","R"))
```

```python
movies_df.groupby(["genre","content_rating"]).get_group(("Crime","R")).sort_values("star_rating")
```

```python
movies_df.groupby(["genre","content_rating"]).get_group(("Crime","R")).sort_values("star_rating",ascending=False)
```
