### Creating DataFrame using Dict,List,Array,Series.....

```python
DictClass = {
    "StudentName" : ["Manisha","Leenata","Abhishek","Ajay","Akash"],
    "StudentAge" : [25,26,27,28,29]
}
```

```python
print(DictClass)
```

```python
df = pd.DataFrame(DictClass)
df
```

```python
arr = np.random.randint(1,50,(5,5))
arr
```

```python
pd.DataFrame(arr)
```

```python
pd.DataFrame(arr,columns=["FirstCol","SecondCol","ThirdCol","FourthCol","FifthCol"])
```

```python
pd.DataFrame(arr,columns=list("ABCDE"))
```

```python
pd.DataFrame(
    arr,
    columns=["FirstCol","SecondCol","ThirdCol","FourthCol","FifthCol"],
    index=["FirstIndex","SecondIndex","ThirdIndex","FourthIndex","FifthIndex"]
)
```

```python
pd.DataFrame(arr,columns=list("ABCDE"),index=list("PQRST"))
```
