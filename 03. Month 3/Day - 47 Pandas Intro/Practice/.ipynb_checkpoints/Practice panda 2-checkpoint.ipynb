## Creating DF with Dictionary

```python
import pandas as pd
import numpy as np
import datetime
```

```python
student_info = {
    "StudentName" : ["Ramesh", "Suresh", "Challo", "Hello", "Bollo"],
    "StudentAge" : [20,30,33,25,24]
}
```
