# 8. insert()
- df.insert(index_number,col_name, value, allow_duplicates = True/False)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame()
df
```

```python
df.insert(0,"Col1",[1,2,3,4,5])
```

```python
df
```

```python
df.insert(0,"Col2",np.array([1,2,3,4,5]) ** 2)
df
```

```python
df.insert(1,"Col3",np.array([1,2,3,4,5]) ** 3)
df
```

```python
df.insert(3,"Col4",150)
df
```

```python
df.insert(4,"Col4",15000)
df
```

```python
df
```

```python
df["Col5"] = df["Col1"].apply(lambda x : x + 50)
df
```

```python
df["Col6"] = df["Col1"] + df["Col2"]
df
```

```python
df["Col7"] = (df["Col1"] + df["Col2"])//2
df
```

```python
df["Col8"] = "DS"
```

```python
df
```
