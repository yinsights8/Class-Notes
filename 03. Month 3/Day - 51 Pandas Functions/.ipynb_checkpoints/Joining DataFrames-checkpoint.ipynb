# 2. concat()
- pd.concat([df1,df2.....dfn],axis=0,ignore_index=True,join="outer")

```python
df1 = pd.DataFrame(
    {
        "Names" : ["Ajay","Suraj","Ganesh","Sanket"],
        "Age" : [23,54,34,25],
        "Location" : ["Pune","Mumbai","Delhi","Chennai"]
    }
)
df1
```

```python
df2 = pd.DataFrame(
    {
        "Names" : ["Akshay","Pushkar","Soham","Rohit"],
        "Age" : [22,24,33,32],
        "Location" : ["Pune","Mumbai","Delhi","Chennai"]
    }
)
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=1)
```
