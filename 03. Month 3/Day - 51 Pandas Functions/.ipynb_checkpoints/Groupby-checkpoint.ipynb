```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("movies_data.csv")
df.head()
```

```python
df.groupby("genre")
```

```python
df.groupby("genre").first()
```

```python
df.groupby("genre").get_group("Western")
```

```python
df.groupby("genre").get_group("Fantasy")
```

```python
df.groupby("genre").get_group("Fantasy")
df.reset_index()
```

```python
df.groupby("genre").get_group("Fantasy")
df.reset_index(drop=True)
```

```python
df['content_rating'].unique()
```

```python
df['content_rating'].nunique()
```

```python
df.groupby("content_rating").first()
```

```python
df['content_rating'].value_counts()
```

```python
df.groupby("content_rating").get_group("GP")
```

```python
df.groupby("content_rating").get_group("PG")
```

```python
 # For multiple categories pass a tuple of categories in get group
df.groupby(["content_rating","genre"]).get_group(("PG","Action"))
```

```python
df.groupby(["content_rating","genre"]).get_group(("PG","Action","Drama"))
```

```python
df.groupby(["content_rating","genre"]).get_group(("PG","Drama"))
```

```python
df.groupby(["content_rating","genre"]).get_group(("PG","Drama")).sort_values("star_rating")
```

```python
df.groupby(["content_rating","genre"]).get_group(("PG","Drama")).sort_values("star_rating",ascending=False)
```

```python
df.groupby(["content_rating","genre"]).get_group(("PG","Drama")).sort_values("duration",ascending=False)
```
