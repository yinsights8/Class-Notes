# Joining DF

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(
    {
        "Names" : ["Akash","Samay","Gaurav","Sham"],
        "Age" : [30,29,25,35],
        "Location" : ["Pune","Mumbai","Delhi","Chennai"],
        "Salary" : [30000,50000,60000,40000]
    }
    ,index = list("ABCD")
)
df1
```

```python
df2 = pd.DataFrame(
    {
        "Names" : ["Akshay","Suresh","Ajinkya","Sham"],
        "Age" : [25,27,85,35],
        "Location" : ["Pune","Mumbai","Delhi","Chennai"],
        "Designation" : ["Python Developer","Data Scientist","Java Developer","Software Engineer"]
    }
    ,index = list("CDEF")
)
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],axis=1,join="outer")
```

```python
pd.concat([df1,df2],axis=1,join="inner")
```

```python
pd.concat([df1,df2],axis=0,join="inner")
```

```python
pd.concat([df1,df2],axis=0,join="outer")
```

```python
pd.concat([df1,df2],axis=0,join="inner")
```

### merge(left_df,right_df,how="inner")

```python
df1 = pd.DataFrame(
    {
        "Names" : ["Akash","Samay","Gaurav","Sham"],
        "Age" : [30,29,25,35],
        "Location" : ["Pune","Mumbai","Delhi","Chennai"],
        "Salary" : [30000,50000,60000,40000]
    }
    ,index = list("ABCD")
)
df1
```

```python
df2 = pd.DataFrame(
    {
        "Names" : ["Akash","Suresh","Ajinkya","Sham"],
        "Age" : [30,27,85,35],
        "Location" : ["Pune","Mumbai","Delhi","Chennai"],
        "Designation" : ["Python Developer","Data Scientist","Java Developer","Software Engineer"]
    }
    ,index = list("CDEF")
)
df2
```

```python
pd.merge(df1,df2,how="inner")
```

```python
pd.merge(df1,df2,how="left")
```

```python
pd.merge(df1,df2,how="right")
```

```python
pd.merge(df1,df2,on="Names",how="outer")
```

```python
pd.merge(df1,df2,on="Names",how="inner")
```

### df1.join(df2, join = "left)

```python
df1 = pd.DataFrame(
    {
        "Col1" : [1,2,3],
        "Col2" : [10,20,30]
    },
    index = list("ABC")
)
df1
```

```python
df2 = pd.DataFrame(
    {
        "Col3" : [100,200,300],
        "Col4" : [1000,2000,3000]
    },
    index = list("ACD")
)
df2
```

```python
df1.join(df2,how="inner")
```

```python
df1.join(df2,how="outer")
```

```python
df1.join(df2,how="left")
```

```python
df1.join(df2,how="right")
```

```python
df1 = pd.DataFrame(
    {
        "Col1" : [1,2,3,3,3,3],
        "Col2" : [10,20,30,30,30,30]
    },
    index = list("ABCDEF")
)
df1
```

```python
df1.drop_duplicates()
```

```python
df1.drop_duplicates(keep="first",ignore_index=True,inplace=True)
```
