# Concat Dataframe

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "Name" : ["Kabir", "Prajwal", "Rushikesh", "Ritik"],
    "Age" : [21,23,24,22],
    "Locattion" : ["Pune", "Mumbai", "Banglore", "Nagpur"],
    "Salary" : [200000,300000,500000,400000]
    
}, 
    index=["A", "B", "C", "D"]

)
```

```python
df1
```

```python
df2 = pd.DataFrame({
    "Name" : ["Kabir", "Ranveer", "Rushikesh", "Kartik"],
    "Age" : [21,25,24,23],
    "Locattion" : ["Pune", "Maisore", "Banglore", "Jalnapur"],
    "Designation" : ["Python Developer","Data Scientist","Java Developer","Software Engineer"]
},
    index = list("CDEF")

)
```

```python
df2
```

```python
concat_1 = pd.concat([df1,df2])
concat_1
```

```python
concat_2 = pd.concat([df2,df1])
concat_2
```

## If any columns are INNER or intersection or common in more than two columnsLables or indexLables

```python
display(df1)
display(df2)
```

```python
cocat_axis_1 = pd.concat([df1,df2],axis=1)
cocat_axis_1
```

## 1. join = "inner"

```python
inner_concat_1 = pd.concat([df1, df2],axis=0,join="inner")
inner_concat_1
```

```python
inner_concat_1 = pd.concat([df1, df2],axis=1,join="inner")
inner_concat_1
```

## 2. join = "outer"

```python
cocat_axis_1
```

```python
concat_outer_1 = pd.concat([df1, df2],axis=1, join="outer")
concat_outer_1
```
