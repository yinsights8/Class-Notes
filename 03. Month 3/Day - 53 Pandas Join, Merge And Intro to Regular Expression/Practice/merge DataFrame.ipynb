## merging dataframe
- Syntax
    - merge(left_df, right_df, how="inner" / "outer")

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "Name" : ["Kabir", "Prajwal", "Rushikesh", "Ritik"],
    "Age" : [21,23,24,22],
    "Locattion" : ["Pune", "Mumbai", "Banglore", "Nagpur"],
    "Salary" : [200000,300000,500000,400000]
    
}, 
    index=["A", "B", "C", "D"]

)
```

```python
df2 = pd.DataFrame({
    "Name" : ["Kabir", "Ranveer", "Rushikesh", "Kartik"],
    "Age" : [21,25,24,23],
    "Locattion" : ["Pune", "Maisore", "Banglore", "Jalnapur"],
    "Designation" : ["Python Developer","Data Scientist","Java Developer","Software Engineer"]
},
    index = list("CDEF")

)
```

```python
display("df1",df1)
display("df2",df2)
```

```python
pd.merge(df1,df2, how="inner")
```

```python
pd.merge(df1,df2, how="outer")
```

```python
pd.merge(df1,df2, how="inner", on="Name")
```

```python
display(df1,df2)
```

```python
pd.merge(left=df1, right=df2, how="outer", on="Name")
```

```python
pd.merge(left=df1, right=df2, how="outer", on="Age")
```

```python
pd.merge(left=df1, right=df2, how="outer", on="Designation")
```

```python
pd.merge(left=df1, right=df2, how="outer", on="Salary")
```

```python
display(df1, df2)
```

```python
pd.merge(left=df1, right=df2, how="outer", on="Locattion")
```

```python
pd.merge(left=df1, right=df2, how="inner", on="Locattion")
```

```python
pd.merge(left=df1, right=df2, how="outer")
```

```python
pd.merge(left=df1, right=df2, left_index=True, right_index=True, how="outer")
```

```python
pd.merge(left=df1, right=df2, left_index=True,right_index=True, how="outer")
```
