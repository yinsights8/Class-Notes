```python
import pandas as pd
```

## Join Dataframe
- Syntax
    - df_name.join(other, on, how, lsuffix, rsuffix, sort)

```python
df1 = pd.DataFrame({
    "Col1" : [1,2,3,4],
    "Col2" : [10,20,30,40]
},
    index = list("ABCD")
)
df1
```

```python
df2 = pd.DataFrame({
    "Col3" : [100,200,300,400],
    "Col4" : [1000,2000,3000,4000]
},
    index = list("CDEF")
)
df2 
```

```python
## C & D index rows are common in both the dataframes
df1.join(df2, how="inner")
```

```python
df1.join(df2,how="outer")
```

```python
df2.join(df1,how="inner")
```

```python
df2.join(df1,how="outer")
```

### join(object, on, how="")

```python
display("df1",df1)
display("df2",df2)
```

```python
df1.join(df2, how="left")
```

```python
df1.join(df2,how="right")
```
