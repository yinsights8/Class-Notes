```python
x_df = pd.DataFrame(Digits.data, columns = Digits.feature_names)
```

```python
x_df
```

```python
y = pd.Series(Digits.target)
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_df,y,test_size=0.2,random_state=1,stratify=y)
```

```python
y_test.value_counts()
```

# Train Model

```python
logistic_model = LogisticRegression(multi_class='ovr')
logistic_model.fit(x_train,y_train)
```

```python
y_pred = logistic_model.predict(x_test)
y_pred_train = logistic_model.predict(x_train)
```

```python
testing_accuracy = accuracy_score(y_test,y_pred)
print('Testing data accuracy :',testing_accuracy)

training_accuracy = accuracy_score(y_train,y_pred_train)
print('Training data accuracy :',training_accuracy)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
multilabel_confusion_matrix(y_test,y_pred)
```

```python
clf_report = classification_report(y_test,y_pred)
print(clf_report)
```
