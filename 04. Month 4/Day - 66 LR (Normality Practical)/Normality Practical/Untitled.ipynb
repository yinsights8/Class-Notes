```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dataset = pd.read_csv("Iris.csv")
dataset.head()
```

```python
sns.histplot(dataset.iloc[:, 1:],kde=True)
```

```python
sns.displot(dataset.iloc[:, 1:],kde=True)
```

```python
sns.histplot(dataset["SepalLengthCm"], kde=True)
```

```python
sns.displot(dataset["SepalLengthCm"], kde=True)
```

```python
sns.displot(dataset["SepalWidthCm"], kde=True)
```

```python
np.mean(dataset["SepalWidthCm"])
```

```python
np.std(dataset["SepalWidthCm"])
```

```python
sns.displot(dataset["PetalLengthCm"], kde=True)
mean = np.mean(dataset["PetalLengthCm"])
std = np.std(dataset["PetalLengthCm"])
print("mean:- ",mean)
print("standard deviation:- ",std)
```

```python
sns.histplot(dataset["Species"], kde=True)
```

```python
sns.displot(dataset["Species"], kde=True)
```
